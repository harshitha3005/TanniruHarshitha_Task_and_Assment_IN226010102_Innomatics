```python
# Problem 1: Post Engagement Analyzer
def analyze_engagement(likes_list):

    total_likes = sum(likes_list)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Test Case
likes = [200, 300, 150, 400]
total, status = analyze_engagement(likes)

print("Total Likes:", total)
print("Post Status:", status)
```

```python
# Problem 2: Medicine Stock Alert System
def medicine_stock_check(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    return stock, status


# Test Case
stock, status = medicine_stock_check(6)
print("Medicine Stock:", stock)
print("Status:", status)
```

```python
# Problem 3: Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):

    average = sum(rainfall_data) / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Test Case
avg, status = rainfall_checker([60, 80, 75, 70], 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

```python
# Problem 4: Duplicate Account Detection
def detect_duplicates(usernames):

    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    return result


# Test Case
result = detect_duplicates(["user1", "user2", "user3", "user1"])
print("Duplicate Accounts Found:", result)
```

```python
# Problem 5: Appointment Eligibility Checker
def check_appointment_eligibility(age):

    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return age, status


# Test Case
age, status = check_appointment_eligibility(21)
print("Patient Age:", age)
print("Eligibility Status:", status)
```

```python
# Problem 6: Premium Crop Price Filter
def filter_premium_crops(prices):

    premium = [price for price in prices if price > 2000]
    return premium


# Test Case
premium = filter_premium_crops([1500, 2500, 1800, 3200])
print("Premium Crops:", premium)
```

```python
# Problem 7: Application Health Checker
def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return errors, status


# Test Case
errors, status = application_health(7)
print("Error Count:", errors)
print("System Status:", status)
```

```python
# Problem 8: Daily Transaction Limit Checker
def transaction_checker(amount):

    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


# Test Case
amount, status = transaction_checker(60000)
print("Transaction Amount:", amount)
print("Transaction Status:", status)
```

```python
# Problem 9: Attendance Eligibility System
def attendance_checker(attendance_list):

    total_classes = len(attendance_list)
    attended = sum(attendance_list)

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Test Case
percentage, eligibility = attendance_checker([1, 1, 0, 1, 1])
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Test Case
bill, usage = electricity_bill(250)
print("Total Bill:", bill)
print("Usage Classification:", usage)
```
