```python
# Problem 1: Employee Performance Bonus Eligibility

# Given employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Finding the highest score
highest_score = max(employees.values())

# Step 2: Finding all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Displaying the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Step 1: Converting to lowercase
query = query.lower()

# Step 2: Removing punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Splitting words
words = query.split()

# Step 4: Counting frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Step 5: Displaying only words with frequency > 1
repeated_words = {word: count for word, count in word_count.items() if count > 1}

print(repeated_words)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Identifying even readings and store (index, value)
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

# Step 2: Displaying result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Counting domain occurrences
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculating percentage usage
total_emails = len(emails)

print("Email Domain Usage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculating average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detecting days more than 30% above average
print("Sales Spike Days:")

for index, value in enumerate(sales, start=1):
    if value > average_sales + (0.30 * average_sales):
        print(f"Day {index}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Counting occurrences
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Displaying duplicates only
print("Duplicate User IDs:")
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
